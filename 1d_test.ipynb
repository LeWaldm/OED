```python
import data_utils as dutils
import numpy as np
import torch
import matplotlib.pyplot as plt
import model_utils as mutils
import distributions as dist
import pandas as pd
import seaborn as sns
%load_ext autoreload
%autoreload 2
```

```python


EPSILON = 1e-6

class Tissue_1d(dutils.Experimenter):
    def __init__(self):
        super().__init__()
        self.center = torch.distributions.Uniform(0,1).sample((1,))
        print(f'center: {self.center}')
        self.radius = 0.5
        self.n_plateau = 1
        self.plateau_fct = lambda x: torch.exp(-((x-self.center)/self.radius)**(2*self.n_plateau)).clip(0,1)
        
        # plotting
        self.X = torch.linspace(0,1,100)
        y = self.plateau_fct(self.X)
        plt.figure(figsize=(3,1))
        plt.plot(self.X,y)
        plt.show()
        self.y = (y > 0.5).int()

    def get_initial_design_params(self):
        return {'x_loc':torch.tensor(0.5)}
    
    def params2design(self, design_params):
        design = design_params['x_loc'].reshape((1,))
        y = (self.plateau_fct(design_params['x_loc']) > 0.5).int()
        return y, design
    
    def execute_design(self, design_params):
        y,x = self.params2design(design_params)
        self.obs_data['y'].append(y.detach().clone())
        self.obs_data['design'].append(x.detach().clone())
        self.obs_data['design_params'].append(
            {k:v.detach().clone() for k,v in design_params.items()})
        
    def get_eval_data(self, exclude_obs=False):
        return self.y, self.X
    
    def verbose_designs(self,
            design_eval=None,
            pred_probs=None,
            metric_values=[],
            metric_names=[],
    ):
        plt.subplots(2,1, figsize=(6,4))

        plt.subplot(2,1,1)
        plt.plot(design_eval, pred_probs)
        plt.ylim([0,1])

        plt.subplot(2,1,2)
        assert len(metric_values) == len(metric_names)
        df = pd.DataFrame(metric_values, 
                            index=metric_names, 
                            columns=range(len(metric_values[0]))).transpose()
        df['number designs'] = df.index + 1
        for i in range(len(metric_values)):
            sns.lineplot(data=df, 
                            y=metric_names[i], 
                            x='number designs', 
                            label=metric_names[i],
                            markers=True)
        plt.title('Metrics of predictions')

class Prior(dist.Distr):
    def __init__(self, 
                 mu = torch.tensor([0.,0.,0.]),
                 std = torch.tensor([1.,1.,1.]),
    ):
        super().__init__()
        self.params = {}
        self.params['mu'] = mu
        self.params['std'] = std
        self.params_constraints['std'] = 'positive'

        self.reparam_trick = True
        self.min_log = torch.log(torch.tensor(EPSILON))

    def log_probs(self,thetas):
        r = thetas[:,0]
        c = thetas[:,1]
        temp = thetas[:,2]
        mu = self.params['mu']
        std = self.params['std']

        prob_r = torch.distributions.Normal(
            loc=mu[0], scale=std[0]
        ).log_prob(torch.log(r))

        prob_c = torch.distributions.Normal(
            loc=mu[1], scale=std[1]
        ).log_prob(c)

        prob_temp = torch.distributions.Normal(
            loc=mu[2], scale=std[2]
        ).log_prob(torch.log(temp))

        res = (prob_c + prob_r + prob_temp).clamp(self.min_log)
        return res

    def sample(self, n):
        mu = self.params['mu']
        std = self.params['std']

        rvs = torch.randn((n,mu.shape[0])) * std.reshape((1,-1)) + mu
        return torch.hstack((
            torch.exp(rvs[:,0]).reshape((-1,1)), 
            rvs[:,1].reshape((-1,1)),
            torch.exp(rvs[:,2]).reshape((-1,1)) ))

    def predict_mle(self):
        mu = self.params['mu']
        out = torch.hstack((torch.exp(mu[0]), mu[1], torch.exp(mu[2]) ))\
            .reshape((1,-1))
        return out
    


class Predictive(dist.Conditional_distr):
    """
    designs: nx(npoints*2)
    ys: nx(npoints)
    """

    def __init__(self):
        super().__init__()
        self.params = {}
        self.reparam_trick = False
        self.eps = 1e-8
    
    def _parse_params(self,thetas):
        r = thetas[:,0]
        c = thetas[:,1:]
        return c,r

    def log_probs(self, ys, design, thetas):

        c,r = self._parse_params(thetas)
        n = ys.shape[0]
        npoints = design.shape[0]
        r = torch.repeat_interleave(r, repeats=npoints, dim=0)
        c = torch.repeat_interleave(c, repeats=npoints, dim=0)
        design = design.repeat(n,1).reshape((-1,))

        logits = c[:,1] * (torch.abs(design - c[:,0]) - r)
        bern_ps = torch.sigmoid(-logits).clamp(self.eps, 1-self.eps)
        log_probs_points = torch.distributions.Binomial(probs=bern_ps)\
            .log_prob(ys.reshape((-1)))
        log_probs = log_probs_points.reshape((n,npoints))
        log_probs = log_probs.sum(axis=1)
        return log_probs
    
    def sample(self, design, thetas):

        c,r = self._parse_params(thetas)
        t = c.shape[0]
        points = design
        npoints = points.shape[0]
        points = points.repeat(t,1).reshape((-1,))
        c = torch.repeat_interleave(c,npoints,0)
        r = torch.repeat_interleave(r,npoints,0)

        logits = c[:,1] * (torch.abs(points - c[:,0]) - r)
        bern_ps = torch.sigmoid(-logits).clamp(self.eps, 1-self.eps)
        samples = torch.distributions.Binomial(total_count=1,probs=bern_ps)\
            .sample(torch.tensor([1]))\
            .reshape((t,-1))
        return samples
    
    def predict_mle(self, design, thetas):
        c,r = self._parse_params(thetas)
        logits = c[:,1] * (torch.abs(design - c[:,0]) - r)
        print(logits)
        bern_ps = torch.sigmoid(-logits).clamp(self.eps, 1-self.eps)
        return bern_ps

# experiment
tissue = Tissue_1d()
prior = Prior()
predictive = Predictive()

# optimization
eig_method = mutils.eig_PCE
eig_optim_method = mutils.eig_cont_optim
n_designs = 4
# optim_args = {'x_loc': ('minmax', torch.tensor(0.0), torch.tensor(1.0))}

# main execution
experiment, thetas_MAP, metric_values, metric_names = mutils.OED_fit(
    tissue, prior,predictive,n_designs,eig_method,eig_optim_method,
    print_every=50, verbose='plot',

    # eig optim
    n_steps_optim = 100,

    # eig method
    n_outer = 500,
    n_inner = 100,

    # variaional inference
    n_steps_varinf=500,
)

tissue.obs_data
```
